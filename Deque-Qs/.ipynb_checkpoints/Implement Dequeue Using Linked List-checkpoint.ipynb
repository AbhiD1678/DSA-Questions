```python
#User function Template for python3

'''
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None

class Deque:
    def __init__(self):
        self.front=None
        self.rear=None
        
        
given front and rear of deque in every operation,
return updated front and rear in each operation
'''


def insertFront(front,rear,data):
    # return: (front, rear)
    #code here
    temp=Node(data)
    if front==None:
        front=temp
        rear=temp
    else:
        temp.next=front
        front.prev=temp
    front=temp
    return front,rear
    
def insertRear(front, rear, data):
    # return: (front, rear)
    temp = Node(data)
    if rear is None:  # If deque is empty
        front = temp
        rear=temp
    else:
        rear.next = temp
        temp.prev = rear
    rear = temp  # Update rear pointer to the newly inserted node
    return front, rear

def delFront(front, rear):
    # return: (front, rear)
    if front is None:  # If deque is empty, just return front and rear
        return front, rear
    else:
        if front == rear:  # If there's only one element in deque
            front = None
            rear = None
        else:  # More than one element in deque
            front = front.next
            if front:
                front.prev = None
        return front, rear
    
    
def delRear(front, rear):
    # return: (front, rear)
    if rear is None:  # If deque is empty, just return front and rear
        return front, rear
    else:
        if front == rear:  # If there's only one element in deque
            front = None
            rear = None
        else:  # More than one element in deque
            rear = rear.prev
            if rear:
                rear.next = None
        return front, rear
```
