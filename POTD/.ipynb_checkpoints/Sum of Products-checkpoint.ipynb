```python
class Solution:
    def pairAndSum(self, n, arr):
        #code here
         #we initially take sum as 0
        sum=0
        #then we iterate through each bit position
        for i in range(32):
            #set bits refer to number of 1's in binary rep of integer
            #we are going to count the set bits at each position
            #inititally the count will be zero
            count=0
            #now we iterate through the array using j
            for j in arr:
                 # Check if the bit at position i is set (1) in the current element
                if j&(1<<i):
                    # Increment the count if the bit is set
                    count+=1
            #calculate the sum contributed by the bit position
            sum+=count*(count-1)//2*(1<<i)
        return sum
```

```python
class Solution:
    def pairAndSum(self, n, arr):
        sum_ = 0
        for i in range(n):
            for j in range(i + 1, n):
                sum_ += (arr[i] & arr[j])
        return sum_
```

```python
#Back-end complete function Template for Python 3

class Solution:

    #Function to find the sum of sum pair bitwise AND and sum.
    def pairAndSum(self,n,arr):
        ans=0

        #iterating over the bits of a number.
        for i in range(32):
            k=0

            #iterating over the array elements.
            for j in arr:

                #checking if the i-th bit of j is set.
                if (j&(1<<i)):
                    k+=1

            #calculating the sum of pair-wise bitwise AND and sum.
            ans+=(1<<i)*(k*(k-1)//2)

        #returning the ans.
        return ans
```
