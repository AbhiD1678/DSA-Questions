```python

import itertools
class Solution:
    def isAdditiveSequence(self, n):
        l = len(n)
        
        # Loop through all possible combinations of dividing the sequence
        for i, j in itertools.combinations(range(1, l), 2):
            
            # Divide the sequence into three parts
            a, b = n[:i], n[i:j]
            
            # Check if the second part is a valid number (no leading zeros)
            if b != str(int(b)):
                continue
            
            # Continue checking if the sequence is additive
            while j < l:
                
                # Calculate the sum of previous two numbers in the sequence
                c = str(int(a) + int(b))
                
                # Check if the sum matches the next part of the sequence
                if not n.startswith(c, j):
                    break
                
                # Update the pointers and continue checking
                j += len(c)
                a, b = b, c
            
            # If the entire sequence is matched, return 1 (True)
            if j == l:
                return 1
        
        # If no additive sequence is found, return 0 (False)
        return 0
```
