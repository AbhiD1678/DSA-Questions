```python
from collections import Counter

class Solution:
    def isEularCircuitExist(self, v, adj):
        l = []
        for i in adj:
            l += i
        c = Counter(l)
        for i in c.keys():
            if c[i] % 2 == 1:
                return 0
        return 1

```

```python
#Back-end complete function Template for Python 3

class Solution:
    def dfs(self, u, adj, used, eu, ev, ans):
        # Perform depth-first search to find the Eulerian circuit
        while(len(adj[u])):
            e = adj[u].pop()
            # If the edge has already been used, skip it
            if(used[e]):
                continue
            used[u] = True
            # Recursively visit the next node in the circuit
            self.dfs(eu[e] ^ ev[e] ^ u, adj, used, eu, ev, ans)
            # Add the current node to the answer list
            ans.append(u)
    
    def isEularCircuitExist(self, v, adj):
        st = set(tuple([-1,-1]))
        eu = []
        ev = []
        ans = []
        used = [False for i in range(v)]
        # Generate the list of edges and check if there are any duplicates
        for i in range(v):
            for j in range(len(adj[i])):
                u = i
                k = adj[i][j]
                if(len(st) and (tuple([u, k]) in st or tuple([k, u]) in st)):
                    continue
                eu.append(u)
                ev.append(k)
                st.add(tuple([u, k]))
                
        # Create a new adjacency list to reflect the reversed edges
        adj1 = [[] for i in range(v)]
        for i in range(len(eu)):
            adj1[eu[i]].append(i)
            adj1[ev[i]].append(i)
        # Check if the degree of each node is even
        for i in range(v):
            if(len(adj[i]) & 1):
                return False
        # Start the DFS from node 0
        ans.append(0)
        self.dfs(0, adj1, used, eu, ev, ans)
        # Check if the number of visited nodes matches the number of edges
        if(len(ans) ^ len(eu) + 1):
            return False
        return True
```
