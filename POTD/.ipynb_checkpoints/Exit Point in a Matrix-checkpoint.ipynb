```python
#User function Template for python3

class Solution:
	def FindExitPoint(self, n, m, matrix):
		# Code here
		def is_valid(x,y):


            if 0<=x < m and 0 <= y < n:

                return True


            else:

                return False


        def next_move(d, x, y):


            if d == 0:

                return (x+1, y)

            elif d ==1:

                return (x, y+1)

            elif d==2:

                return (x-1, y)

            elif d ==3:

                return (x, y-1)


            
        d = 0

        x = 0

        y = 0





        while True:

             d = (d + matrix[y][x])% 4

             nx, ny = next_move(d, x, y)

             if is_valid(nx,ny):

                  matrix[y][x] = 0

                  x, y = nx, ny

             else:

                 return (y,x)
```

```python
class Solution:

    def FindExitPoint(self, n, m, matrix):
        # Initialize starting position (0, 0)
        i = 0
        j = 0
        # Initialize direction to 1 (1: right, 2: down, 3: left, 4: up)
        dir = 1

        # Loop until we reach the boundary of the matrix
        while 0 <= i < n and 0 <= j < m:
            # If current cell contains 1, change direction and mark the cell as visited (0)
            if matrix[i][j] == 1:
                matrix[i][j] = 0
                dir = (dir % 4) + 1

            # Move to the next cell based on the direction
            if dir == 1:
                j += 1
            elif dir == 2:
                i += 1
            elif dir == 3:
                j -= 1
            elif dir == 4:
                i -= 1

        # Compensate for the last move (backtrack)
        if dir == 1:
            j -= 1
        elif dir == 2:
            i -= 1
        elif dir == 3:
            j += 1
        elif dir == 4:
            i += 1

        # Return the coordinates of the exit point
        return [i, j]
```
