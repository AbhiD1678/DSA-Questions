```python
'''
Given a string s which contains only lower alphabetic characters, check if it is possible to remove at most one character from this string in such a way that frequency of each distinct character becomes same in the string. Return true if it is possible to do else return false.

Note: The driver code print 1 if the value returned is true, otherwise 0.

Example 1:

Input:
s = "xyyz"
Output: 
1 
Explanation: 
Removing one 'y' will make frequency of each character to be 1.
Example 2:

Input:
s = "xxxxyyzz"
Output: 
0
Explanation: 
Frequency can not be made same by removing at most one character.
Your Task:  
You dont need to read input or print anything. Complete the function sameFreq() which takes a string as input parameter and returns a boolean value denoting if same frequency is possible or not.

Expected Time Complexity: O(|s|) 
Expected Auxiliary Space: O(1)
'''
```

```python
import collections
from collections import Counter
import math
class Solution:
    def sameFreq(self, s):
        # code here
        def check(c):
            mn,mx = math.inf,0
            mn_cnt,mx_cnt = 0,0
            for i in c:
                mn = min(mn,c[i])
                mx = max(mx,c[i])
                
            for i in c:
                if mn == c[i]:
                    mn_cnt += 1
                elif mx == c[i]:
                    mx_cnt += 1
            x = len(c)
            if (x - mn_cnt == 0) or (x - mn_cnt == 1 and mx - mn == 1) or (mn_cnt == 1 and x - mx_cnt == 1 and mn == 1):
                return True
            return False
```

```python
class Solution:
    def sameFreq(self, s):
        # code here
        freq = {}
        for char in s:
            freq[char] = freq.get(char, 0) + 1
    
        distinct_freq = set(freq.values())
        if len(distinct_freq) == 1:
            return True
        elif len(distinct_freq) == 2:
            min_freq, max_freq = min(distinct_freq), max(distinct_freq)
            if (min_freq == 1 and list(freq.values()).count(min_freq) == 1) or max_freq - min_freq == 1:
                if list(freq.values()).count(max_freq) == 1:
                    return True
                elif min_freq == 1 and list(freq.values()).count(min_freq) == 1:
                    return True
        return False
```

```python
class Solution:


	def getIdx(self,ch):
		return ord(ch) - ord('a')


	def allSame(self,freq, N):
		# get first non-zero element
		for i in range(0, N):
			if freq[i] > 0:
				same = freq[i]
				break

		# check equality of each element
		# with variable same
		for j in range(i + 1, N):
			if freq[j] > 0 and freq[j] != same:
				return False

		return True


	# Returns true if we can make all
	# character frequencies same
	def sameFreq(self,str1):
		l = len(str1)

		# fill frequency array
		freq = [0] * 26
		for i in range(0, l):
			freq[self.getIdx(str1[i])] += 1

		# if all frequencies are same,
		# then return true
		if self.allSame(freq, 26):
			return True

		# Try decreasing frequency of all character
		# by one and then check all equality of all
		# non-zero frequencies
		for i in range(0, 26):

			# Check character only if it
			# occurs in str
			if freq[i] > 0:
				freq[i] -= 1

				if self.allSame(freq, 26):
					return True
				freq[i] += 1

		return False
```
