```python
class Solution:    
    ##Complete this function
    # function to convert a given Gray equivalent n to Binary equivalent.
    def grayToBinary(self,n):
        ##Your code here
        res = 0
        
        while n > 0:
            res ^= n
            n >>= 1
            
        return res
```
