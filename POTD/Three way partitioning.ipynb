```python
class Solution:
    # Function to partition the array around the range such 
    # that array is divided into three parts.
    def threeWayPartition(self, array, a, b):
        # code here 
        n = len(array)
        i, j = 0, n - 1
        while i < j:
            if array[i] < a:
                i += 1
                continue
            if array[j] >= a:
                j -= 1
                continue
            array[i], array[j] = array[j], array[i]
            i += 1
            j -= 1
        j = n - 1
        while array[i] < a:
            i += 1
        while i < j:
            if array[i] >= a and array[i] <= b:
                i += 1
                continue
            if array[j] > b:
                j -= 1
                continue
            array[i], array[j] = array[j], array[i]
            i += 1
            j -= 1
```
