```python
class Solution:
    def findMaxSum(self,n,m,mat):
        #code here
        if n<3 or m<3:
            return -1
        i , j = 1, 1
        s = 0
        while i <n-1:
            while j<m-1:
                curr = mat[i][j]+mat[i-1][j]+mat[i-1][j-1]+mat[i-1][j+1]+mat[i+1][j-1]+mat[i+1][j]+mat[i+1][j+1]
                s = max(s,curr)
                j += 1
            j = 1
            i += 1
                
        return s

```

```python
class Solution:
    def hoverglassSum(self, r, c, Mat):
        sum_val = Mat[r + 1][c + 1]
        for i in range(3):
            sum_val += Mat[r][c + i]
            sum_val += Mat[r + 2][c + i]
        return sum_val
    
    def findMaxSum(self, N, M, Mat):
        max_sum = -1
        for i in range(N - 2):
            for j in range(M - 2):
                max_sum = max(max_sum, self.hoverglassSum(i, j, Mat))
        return max_sum
```

```python
#Back-end complete function Template for Python 3

class Solution:
    def findMaxSum(self,N,M,Mat):
        if N<3 or M<3:
            return -1
        # Here loop runs (R-2)*(C-2) times considering
        # different top left cells of hour glasses.
        
        max_sum=-1
        for i in range(N-2):
            for j in range(M-2):
                # Considering mat[i][j] as top left cell of
                # hour glass.
                sum = (Mat[i][j]+Mat[i][j+1]+Mat[i][j+2])+(Mat[i+1][j+1])+(Mat[i+2][j]+Mat[i+2][j+1]+Mat[i+2][j+2])
                # If previous sum is less then current sum then
                # update new sum in max_sum
                max_sum=max(max_sum,sum)
        return max_sum  
```
