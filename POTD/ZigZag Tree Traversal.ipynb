```python
'''
Given a binary tree with n nodes. Find the zig-zag level order traversal of the binary tree.

Example 1:

Input:
        1
      /   \
     2    3
    / \    /   \
   4   5 6   7
Output:
1 3 2 4 5 6 7
Example 2:

Input:
           7
        /     \
       9      7
     /  \      /   
    8   8  6     
   /  \
  10  9 
Output:
7 7 9 8 8 6 9 10 
Your Task:
You don't need to read input or print anything. Your task is to complete the function zigZagTraversal() which takes the root node of the Binary Tree as its input and returns a list containing the node values as they appear in the zig-zag level-order traversal of the tree.

Expected Time Complexity: O(n).
Expected Auxiliary Space: O(n).

Constraints:
1 <= n <= 105


'''
```

```python
class Solution:
    #Function to store the zig zag order traversal of tree in a list.
    def zigZagTraversal(self, root):
        queue = [[root]]
        res = []
        check = False
        while len(queue) != 0:
            temp = queue.pop()
            nodes = []
            val = []
            for one in temp:
                val.append(one.data)
                if one.left:
                    nodes.append(one.left)
                if one.right:
                    nodes.append(one.right)
            if len(nodes) > 0:
                queue.append(nodes)
            if check:
                res.extend(val[::-1])
                check = False
                continue
            check = True
            res.extend(val)
            
        return res
```

```python
'''
Expected Approach
Intuition
The idea is to use two stacks to store the order in which the nodes are to be visited in current level and in the next level, if current level is left to right then push nodes in next level from left to right because stack is LIFO (Last-In-First-Out) data structure.

Implementation
If null then return,
Declare two stacks 'currentLevel' and 'nextLevel' to store the nodes.
We also declare variable 'lefttoright' to keep track of the current level order(whether it is left to right or right to left).
Push the root into the currentLevel stack
Check if currentLevel is not empty,
Storing the top element of currentlevel stack in temp and popping it.
If temp is not null, we store the data at temp in list that store the zig-zag traversal.
If lefttoright is true then it stores nodes from left to right else from right to left in nextlevel stack as stack is LIFO so when popped it would come out as right to left.
Again check if currentlevel stack is empty lefttoright is flipped to change the order of storing the nodes and both stacks are swapped.
Repeat this process untill all the levels are traversed,
Return the list storing the traversal. 
Example

 
At first we intialize currentLevel with root i.e. 1, and make lefttoright = true
    Since currentLevel i.e. {1} is not empty we pop 1 and push it in 'res' i.e. {1},
        Store 1's child in nextLevel in left to right order i.e. nextLevel =  {2,3}
        Now current level is empty so swap currentLevel and nextLevel and make lefttoright = false
    As current level i.e. {2,3} is not empty we pop 3 from top, push it in 'res' i.e. {1,3}
       Store 3's child in nextLevel in right to left i.e. nextLevel = {4,5}
       Now pop 2 from currentLevel and push it in 'res' i.e. {1,3,2} 
       Store 2's child in nextLevel in right to left i.e. nextLevel = {4,5,6,7}
       Now current level is empty so swap currentLevel and nextLevel and make lefttoright = true
   Again currentLevel is not empty so we pop 7 and push it in res i.e. {1,3,2,7}
       As 7 has no children, we ,move to 6 and push it in res i.e. {1,3,2,7,6}
       Similarly we do for 5 and 4 respectively and res becomes {1,3,2,7,6,5,4} and currentLevel becomes empty
       Now we swap currentLevel and nextLevel and make lefttoright = false
   As currentLevel is empty, we return res as our output.
Complexity
Time Complexity: O(N), for visiting every node in the tree. 
Space Complexity: O(N)+(N)=O(N), for storing the nodes in two stacks, where N is the number of nodes present in the binary tree.
'''
#Back-end complete function Template for Python 3

class Solution:
    #Function to store the zig zag order traversal of tree in a list.
    def zigZagTraversal(self, root):
 
        res = []
        #if root is null, we return the list.
        if root is None:
            return res
 
        #declaring two stacks to store the current and new level.
        currentLevel = []
        nextLevel = []

        ltr = True
 
        #pushing the root in currentlevel stack.
        currentLevel.append(root)
 
        while len(currentLevel) > 0:
            
            #storing the top element of currentlevel stack in temp and popping it.
            temp = currentLevel.pop(-1)
       
            #if temp is not null, we store the data at temp in list.
            res.append(temp.data)
 
            #if lefttoright is true then it stores nodes from left to right 
		    #else from right to left in nextlevel stack.
            if ltr:
                if temp.left:
                    nextLevel.append(temp.left)
                if temp.right:
                    nextLevel.append(temp.right)
            else:
                if temp.right:
                    nextLevel.append(temp.right)
                if temp.left:
                    nextLevel.append(temp.left)
 
            #if currentlevel stack is empty ltr is flipped to change 
		    #the order of storing the nodes and both stacks are swapped.
            if len(currentLevel) == 0:
            
                ltr = not ltr
                currentLevel, nextLevel = nextLevel, currentLevel
    
        #returning the list.    
        return res   
```

```python
'''
Alternate Approach
Intuition
The idea is to use a deque to solve the problem. Push and pop in different ways depending on if the level is odd or level is even i.e. depending upon if we need to traverse from left to right or from right to left.

Implementation
Initialize a deque 'q' (with root) to store the levels for traversal and a vecter 'res' (with root) to store the results and a l variable 'level' (with 1) to keep track of left to right or right to left .
We check if the dequeue is not empty
If so, we find its size (store it in n) and run a loop from 0 to n,
If level is even, we pop the dequeue from back so that we maintain left to right traversal , else if level is odd we pop it from front to maintain right to left traversal
If level is even we push the nodes into 'q' from front so that in next turn when level becomes odd right to left traversal can be maintained, and we push the nodes value into 'res' in the order that they are visited.
Now increase the level by 1, and perform the above mentioned steps again.
When dequeue becomes empty, return the result vector i.e. 'res'.
Example

 
At first dequeue 'q' is initialzed with {1}, result vector 'res' is initialized with {1} and 'level' = 1
    As the q is not empty and level is odd, we pop 1 from q
        Push back 1's right and and left child into the 'q' making it {3,2} and 'res' making it {1,3,2} respectively
    Now level becomes 2,
        we pop 2 from 'q' and push front its left and right child into 'q' making it {7,6,3} and push back the same into res making it {1,3,2,7,6}
        we pop 3 from 'q' and push front its left and right child into 'q' making it {7,6,5,4} and push back the same into res making it {1,3,2,7,6,5,4}
    Now level becomes 3,
        we pop 7 from 'q' but its leaf node so nothing happens, similarly for node 6,5,and 4.
Finally we return 'res' i.e. {1,3,2,7,6,5,4}
'''
class Solution:
    #Function to store the zig zag order traversal of tree in a list.
    def zigZagTraversal(self, root):
        if not root:
            return []
        
        q = deque()
        res = []
        q.append(root)
        res.append(root.data)
        level = 1
        
        while q:
            n = len(q)
            for i in range(n):
                if level % 2 == 0:
                    temp = q.pop()
                else:
                    temp = q.popleft()
                
                if level % 2 != 0:
                    if temp.right:
                        q.append(temp.right)
                        res.append(temp.right.data)
                    if temp.left:
                        q.append(temp.left)
                        res.append(temp.left.data)
                else:
                    if temp.left:
                        q.appendleft(temp.left)
                        res.append(temp.left.data)
                    if temp.right:
                        q.appendleft(temp.right)
                        res.append(temp.right.data)
            level += 1
        
        return res
'''
Complexity
Time Complexity: O(N), for traversing the binary tree.
Auxiliary Space: O(N), for deque data structure, where N is the number of nodes in the binary tree.
'''
```
