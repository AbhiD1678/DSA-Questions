```python
class Solution:
    def solve(self, root, v):
        if root is None:
            return
        
        self.solve(root.left, v)
        v.append(root.data)
        self.solve(root.right, v)
    
    def absolute_diff(self,root):
        v = []
        self.solve(root, v)
        
        n = len(v)
        mini = float('inf')
        
        for i in range(n - 1):
            for j in range(i + 1, n):
                mini = min(mini, v[j] - v[i])
        
        return mini
```

```python
class Solution:
    def inorderTraverse(self, root, val, min_diff):
        # Inorder traversal of the binary tree
        if root.left != None: # If left child exists
            val, min_diff = self.inorderTraverse(root.left, val, min_diff)
            
        if val>=0: # If val is not the initial value
            min_diff = min(min_diff, root.data-val) # Calculate the difference and update min_diff
        
        val = root.data # Update val to current node's data
        if root.right !=None: # If right child exists
            val, min_diff = self.inorderTraverse(root.right, val, min_diff)
            
        return val, min_diff
    
    def absolute_diff(self, root):
        min_diff = float('inf') # Initialize min_diff to infinity
        val = -1 # Initialize val to -1
        return self.inorderTraverse(root, val, min_diff)[1] # Return the min_diff after completing the inorder traversal
```
